```python
#FIB	Rabbits and Recurrence Relations

def total_conejos(n, k):
    # Casos iniciales para el mes 1 y mes 2
    if n == 1 or n == 2:
        return 1

    # Inicializamos los valores para los dos primeros meses
    f_menos_2 = 1  # Mes 1
    f_menos_1 = 1  # Mes 2
    f_actual = 0

    # Calculamos mes a mes desde el mes 3 hasta n
    for mes in range(3, n + 1):
        f_actual = f_menos_1 + (k * f_menos_2)
        # Actualizamos las variables para el siguiente mes
        f_menos_2 = f_menos_1
        f_menos_1 = f_actual

    return f_actual

# Prueba con los datos del ejemplo (n = 5, k = 3)
n = 34
k = 4
resultado = total_conejos(n, k)
print(f"Resultado para n={n} y k={k}: {resultado}")
```

```python
#GC	Computing GC Content

def calcular_gc(secuencia):
    conteo_gc = secuencia.count('G') + secuencia.count('C')
    porcentaje = (conteo_gc / len(secuencia)) * 100
    return porcentaje

# 1. Leemos el archivo y creamos la lista
from google.colab import drive
drive.mount('/content/drive')
with open("/content/drive/MyDrive/rosalind_gc.txt", "r") as archivo:
    lineas = archivo.readlines()

# 2. Creamos la tabla vacía
secuencias = {}
id_actual = ""

# 3. Rellenamos la tabla línea por línea
for linea in lineas:
    linea = linea.strip()
    if linea.startswith('>'):
        id_actual = linea[1:]
        secuencias[id_actual] = ""
    else:
        secuencias[id_actual] += linea
for id_dna, secuencia_completa in secuencias.items():
    # Usamos la función que creamos arriba
    porcentaje_calculado = calcular_gc(secuencia_completa)

    # Transformamos la fila: ahora guarda [secuencia, porcentaje]
    secuencias[id_dna] = [secuencia_completa, porcentaje_calculado]
# 1. Creamos las variables para anotar al ganador.
# Empezamos con un porcentaje muy bajo (-1.0) para que cualquiera lo supere al inicio.
id_maximo = ""
gc_maximo = -1.0

# 2. Recorremos la tabla fila por fila.
# 'id_dna' es el ID de la fila.
# 'datos' es la lista que contiene: [secuencia, porcentaje]
for id_dna, datos in secuencias.items():

    # Extraemos el porcentaje. Como es el SEGUNDO elemento de la lista,
    # en Python accedemos a él con el índice [1] (recuerda que informática empieza en 0).
    porcentaje_actual = datos[1]

    # 3. Comparamos: ¿Este porcentaje es mayor que el máximo que teníamos anotado?
    if porcentaje_actual > gc_maximo:
        gc_maximo = porcentaje_actual  # Anotamos el nuevo porcentaje más alto
        id_maximo = id_dna             # Anotamos el ID del dueño de ese porcentaje

# 4. Cuando el bucle termina de revisar todas las filas, imprimimos al ganador absoluto.
# Usamos :.6f para que muestre exactamente 6 decimales, como pide Rosalind.
print(id_maximo)
print(f"{gc_maximo:.6f}")
```

```python
#IPRB	Mendel's First Law

def probabilidad_dominante(k, m, n):
    # k = homocigotos dominantes (AA)
    # m = heterocigotos (Aa)
    # n = homocigotos recesivos (aa)

    # Población total
    N = k + m + n
    # Total de parejas ordenadas posibles sin reemplazo
    total_parejas = N * (N - 1)

    # Número de parejas posibles multiplicadas por su probabilidad de dar un descendiente 'aa'
    recesivos_aa_aa = n * (n - 1) * 1.0       # aa x aa -> 100% aa
    recesivos_Aa_aa = (m * n + n * m) * 0.5   # Aa x aa o aa x Aa -> 50% aa
    recesivos_Aa_Aa = m * (m - 1) * 0.25      # Aa x Aa -> 25% aa

    # Suma de todas las probabilidades de obtener recesivos
    prob_recesivo = (recesivos_aa_aa + recesivos_Aa_aa + recesivos_Aa_Aa) / total_parejas

    # La solución es el complemento (1 - prob_recesivo)
    prob_dominante = 1 - prob_recesivo
    return prob_dominante

# Datos del ejemplo
k, m, n = 16, 26, 21
resultado = probabilidad_dominante(k, m, n)

# Mostrar el resultado formateado a 5 decimales
print(f"{resultado:.5f}")
```

```python
#CONS	Consensus and Profile

from collections import Counter

secuencias = []
secuencia_actual = []

from google.colab import drive
drive.mount('/content/drive')
with open("/content/drive/MyDrive/rosalind_cons.txt", "r") as archivo:
    for linea in archivo:
        linea = linea.strip()
        if not linea:
            continue

        if linea.startswith(">"):
            if secuencia_actual:
                secuencias.append("".join(secuencia_actual))
                secuencia_actual = []
        else:
            secuencia_actual.append(linea)

    if secuencia_actual:
        secuencias.append("".join(secuencia_actual))

longitud_secuencia = len(secuencias[0])

# Inicializamos el perfil con listas de ceros para cada base (A, C, G, T)
perfil = {
    'A': [0] * longitud_secuencia,
    'C': [0] * longitud_secuencia,
    'G': [0] * longitud_secuencia,
    'T': [0] * longitud_secuencia
}

secuencia_consenso = []

# Analizamos el archivo de forma vertical (posición por posición)
for i in range(longitud_secuencia):
    # Extraemos la letra en la posición 'i' de todas las secuencias
    columna = [seq[i] for seq in secuencias]
    conteo = Counter(columna)

    # Rellenamos la matriz de perfil con las cantidades encontradas
    for base in ['A', 'C', 'G', 'T']:
        perfil[base][i] = conteo[base]

    # Extraemos la base más común para la secuencia consenso
    # most_common(1)[0][0] obtiene directamente el carácter (ej: 'A')
    base_mas_comun = conteo.most_common(1)[0][0]
    secuencia_consenso.append(base_mas_comun)

# 1. Imprime la secuencia consenso unificada
print("".join(secuencia_consenso))

# 2. Imprime el perfil de cada base separado por espacios
for base in ['A', 'C', 'G', 'T']:
    valores_espaciados = " ".join(map(str, perfil[base]))
    print(f"{base}: {valores_espaciados}")
```
